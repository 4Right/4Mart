```python
import requests
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
import pandas as pd
import matplotlib.font_manager as fm
import urllib.request
from matplotlib import rc
from bs4 import BeautifulSoup
import folium
```

```python
topmart_df = pd.read_csv('./topmart_shop_info_loc.csv', encoding='utf-8')
topmart_df.tail()
```

```python
map = folium.Map(location=[35.2171341,129.1111094],zoom_start=10)

for i in topmart_df.index:
    lt_lat = topmart_df.loc[i,'Latitude']
    lt_long = topmart_df.loc[i,'Longitude']

    name = f"[{topmart_df.loc[i, 'store']}]"
    address = f"{topmart_df.loc[i, 'address']}"
    tel = f"{topmart_df.loc[i, 'tel']}"

    color = 'orange'
    if topmart_df.loc[i,'store']=='탑마트':
        color ='orange'

    
    # popup = folium.Popup(iframe, min_width=200, max_width=200)
    folium.Marker(location=[lt_lat,lt_long],icon=folium.Icon(icon='barcode',color='orange'),popup=f'<pre>{name}\\n 주소: {address}\\n Tel: {tel}</pre>').add_to(map)
    
    
    
#     iframe = location_name+ ":<br> "+ addr 
# popup = folium.Popup(iframe, min_width=200, max_width=200)

# Marker(location=[addr_lat, addr_lon], popup=popup, tooltip=location_name, icon=Icon(color='green', icon='flag')).add_to(m)
map
```
