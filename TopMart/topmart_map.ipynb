```python
import requests
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
import pandas as pd
import matplotlib.font_manager as fm
import urllib.request
from matplotlib import rc
from bs4 import BeautifulSoup
import folium
```

```python
topmart_df = pd.read_csv('./topmart_shop_info_loc.csv', encoding='utf-8')
topmart_df.tail()
```

```python
map = folium.Map(location=[35.2171341,129.1111094],zoom_start=10)

for i in topmart_df.index:
    lt_lat = topmart_df.loc[i,'Latitude']
    lt_long = topmart_df.loc[i,'Longitude']

    title = f"{topmart_df.loc[i, 'store']} - {topmart_df.loc[i, 'address']}"

    color = 'orange'
    if topmart_df.loc[i,'store']=='탑마트':
        color ='orange'

    folium.Marker(location=[lt_lat,lt_long],radius=5,color=color,icon=folium.Icon(icon='barcode',color='orange'),tooltip='탑마트').add_to(map)
map
```
