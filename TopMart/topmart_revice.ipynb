```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./data/topmart_shop_info_loc.csv', index_col=0, encoding='utf-8')
data.head()
```

```python
data.insert(0, 'store_name', '탑마트')
```

```python
data.head()
```

```python
data['name'] = data.apply(lambda r: r['store_name'] + ' ' + r['store'], axis='columns')
data.head()
```

```python
data = data.drop('store_name', axis=1)
```

```python
data = data.drop('store', axis=1)
```

```python
data.head()
```

```python
data = data.rename(columns={'name':'store'})
```

```python
data = data.rename(columns={'phone':'tel'})
```

```python
data.head()
```

```python
data = data[['store', 'address', 'Latitude', 'Longitude', 'tel']]
```

```python
data.head()
```

```python
data.to_csv('topmart_shop_info_loc.csv', index=True)
```
