```python
mart_df=pd.read_csv('./mart2.csv',encoding='utf-8')
```

```python
mart_df.head()
```

```python
adr = pd.DataFrame(mart_df['address'].apply(lambda v:v.split()[:2]).tolist(),columns=('시도','구군'))
adr.head()
```

```python
mart_2=pd.merge(adr,mart_df,how='inner',left_index=True,right_index=True)
mart_2.head()
```

```python
mart_2
```

```python
mart_2['count']=0
mart_2.head()
```

```python
mart_group = pd.DataFrame(mart_2.groupby(['시도','구군','brand'],as_index=False).count())
mart_group
```

```python
mart_group =mart_group.drop(columns=['store','address'])
```
