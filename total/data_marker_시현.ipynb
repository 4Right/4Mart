```python
map = folium.Map(location=[36.3895388, 127.9541354],zoom_start=7)

for n in mart_df2.index: #팝업에 들어갈 텍스트를 지정해준다.
    popup_name = mart_df2['store'][n] + mart_df2['address'][n] + mart_df2['tel'][n]
    if mart_df2['brand'][n] == '롯데마트':
        icon_color = 'lightgreen'
    elif mart_df2['brand'][n] == '이마트':
        icon_color = 'gray'
    elif mart_df2['brand'][n] == '홈플러스':
        icon_color = 'red'
    else:
        icon_color = 'orange'

    folium.Marker([mart_df2['Latitude'][n], mart_df2['Longitude'][n]], popup=popup_name, icon=folium.Icon(color=icon_color)).add_to(map)

map
```
