```python
import base64
map = folium.Map(location=[36.3895388, 127.9541354],zoom_start=10)

for i in top_df.index:
    lotte_lat = lotte_df.loc[i,'Latitude']
    lotte_long = lotte_df.loc[i,'Longitude']
    lotte_name = f"[{lotte_df.loc[i, 'store']}]"
    lotte_address = f"{lotte_df.loc[i, 'address']}"
    lotte_tel = f"{lotte_df.loc[i, 'tel']}"
    
    home_lat = home_df.loc[i,'Latitude']
    home_long = home_df.loc[i,'Longitude']
    home_name = f"[{home_df.loc[i, 'store']}]"
    home_address = f"{home_df.loc[i, 'address']}"
    home_tel = f"{home_df.loc[i, 'tel']}"
    
    top_lat = top_df.loc[i,'Latitude']
    top_long = top_df.loc[i,'Longitude']
    top_name = f"[{top_df.loc[i, 'store']}]"
    top_address = f"{top_df.loc[i, 'address']}"
    top_tel = f"{top_df.loc[i, 'tel']}"
    
    emart_lat = emart_df.loc[i,'Latitude']
    emart_long = emart_df.loc[i,'Longitude']
    emart_name = f"[{emart_df.loc[i, 'store']}]"
    emart_address = f"{emart_df.loc[i, 'address']}"
    emart_tel = f"{emart_df.loc[i, 'tel']}"

    lotte_mart = folium.Marker(location=[lotte_lat,lotte_long],icon=folium.Icon(icon='gift',color='lightgreen'),popup=f'<pre><h4>{lotte_name}</h4> \\n 주소: {lotte_address} \\n Tel: {lotte_tel}</pre>').add_to(map)
    homeplus_mart = folium.Marker(location=[home_lat,home_long],icon=folium.Icon(color='red',icon='home'),popup=f'<pre><h4>{home_name}</h4> \\n 주소: {home_address} \\n Tel: {home_tel}</pre>').add_to(map)
    emart_mart = folium.Marker(location=[emart_lat,emart_long],icon=folium.Icon(color='gray',icon='shopping-cart'),popup=f'<pre><h4>{emart_name}</h4> \\n 주소: {emart_address} \\n Tel: {emart_tel}</pre>').add_to(map)
    top_mart = folium.Marker(location=[top_lat,top_long],icon=folium.Icon(icon='barcode',color='orange'),popup=f'<pre><h4>{top_name}</h4> \\n 주소: {top_address} \\n Tel: {top_tel}</pre>').add_to(map)

map
```
